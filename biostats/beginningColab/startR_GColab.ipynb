## グラフを描いてみる
呪文のように見えるかと思いますが実行するとグラフが現れるはずです。

```r
data <- rnorm(1000, 0, 1)
plot(data, type = "l", col = "blue")
```

一応、解説。

rnorm(1000, 0, 1) で1000個のランダムな数（乱数）を発生させています。その1000個の数の平均が0で、標準偏差（データのばらつきの指標）が1です。
head(データ名）で今回作ったdataの最初の方、tail（data)で最後の方をチラ見することができます。試しにtail（data)と入れてみましょう。自動で候補が表示されるので、そこから選ぶだけです。

```r
head(data)
# ここに tail(...)　を入れてみましょう。
```

このグラフは発生した数字をとりあえず折れ線グラフにしたものです。
実行する度に乱数を発生し直すので、その度にグラフの形が変わります。

せっかくなので別のグラフを。今、発生させた乱数がどんな分布になっているかヒストグラムを描いてみます。（これを見ると、平均が0なのもイメージできるかと思います）

```r
hist(data)
```

このヒストグラムはちょっと荒く地味なので、呪文をつけたすと細かくしたり色をつけたりすることもできます。

```r
hist(data, breaks = 50, col = "pink")
```

## もう少し本格的な統計解析
Rにはデモデータとしてiris（アヤメ）のデータが入っているので、これを使って統計解析をチラ見してみましょう。

（チュートリアルムービーサイト統合TVでも解説があります。 http://togotv.dbcls.jp/20180408.html ）

まず下準備として、よりよいグラフを描くためのライブラリ（=プログラム集）をインストールし呼び出しておきます。

```r
install.packages("ggplot2")
library("ggplot2")
```

データの中身を確認します。

```r
head(iris)
nrow(iris)
```

（全部見るとわかるんですが）3種類のアヤメのがくと花びらの花びらの長さと幅のデータになっています。
- Sepal: 萼片（がくへん）
- Petal: 花びら
- Species: 種名

今回は以下の3種類です。
- Iris setosa 270px-Iris_setosa01.jpg
- Iris versicolor 240px-Blue_Flag,_Ottawa.jpg
- Iris virginica 240px-Iris_virginica_2.jpg

nrowでデータの個数がわかります。（厳密には行数をカウント）。今回は150個です。（各種、50個のデータで3種類）

たとえば、各々の萼片の長さを見るには以下のようにします。
- hist(データ名$列名)

```r
hist(iris$Sepal.Length)
```

そして、この場合、3種のデータそれぞれのがく片の長さの分布の比較は以下のようにします。
  - qplot(X軸, Y軸, data=データ名, geom="グラフの種類")

```r
qplot(Species, Sepal.Length, data=iris, geom="boxplot")
```

ここではqplotを使っていますが、別にいちいち関数名を覚える必要はナシ。
誰かがよさげなのを使っていたら拝借するだけです（今回は統合TVで使っているからコピペしただけ）。今後、こういうグラフが描きたいなーという場合は、ググって見つけるか、以前に書いたものを発掘して使い回すかです。

同様に花びらの幅についてもグラフを描いて確認してみましょう。
（上の部分をコピペして）

```r
qplot(Species, Petal.Width, data=iris, geom="boxplot")
```

統合TVでは、I. setosaとI. versicolorの花びらの幅に統計的な差があるか確認しています。
（統計的な差：差があるのは偶然に起きる範囲か、確率を計算する）

150個のデータがあるのですが、1〜50番目がI. setosa、51〜100番目がI. versicolorになっています。
実際の「統計的な有意差があるか」は、t検定という専門的な方法があり、t.test関数を使います。

```r
t.test(iris$Petal.Width[1:50], iris$Petal.Width[51:100])
```

「統計的有意差」はp-valueの値を見ます。単純には偶然に事象（この場合、幅が違うこと）が起きる確率だと思ってください。今回は、かなーり値が小さい（2.2e-16。つまり10^-16のオーダー）ので 幅が偶然に違う確率は小さい＝幅が違うことには意味がある（有意である）と結論づけられます。
